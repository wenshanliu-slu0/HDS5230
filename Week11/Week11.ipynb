```python
# load the datasize =100 
import pandas as pd
data1= pd.read_csv('dataset/100_.csv')

print(data1.shape)
print(data1.head())
```

```python
# split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data1.drop('outcome', axis=1), data1['outcome'], test_size=0.2, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import time

def time_it():
    start_time = time.time()
    
    # n_estimators = nrounds
    model = XGBClassifier(
        n_estimators=100,               # n_estimators is the number of trees in the model
        early_stopping_rounds=10,        # if 10 rounds are not improved, stop
        eval_metric="logloss",           # monitor the evaluation metric
        use_label_encoder=False,         # close the warning (old version needs)
        verbosity=0                      # close the training process output
    )

    # provide the validation set to enable early stopping
    model.fit(
        X_train, y_train,
        eval_set=[(X_test, y_test)],     # provide the validation set
        verbose=False                    # close the training process output
    )

    # cross-validation evaluation
    cv_scores = cross_val_score(XGBClassifier(n_estimators=model.best_iteration), X_train, y_train, cv=5)
    
    #calculate the time it takes to train the model
    end_time = time.time()

    print("5-fold CV Scores:", cv_scores)
    print("Mean CV Score:", cv_scores.mean())
    print("CV Score Standard Deviation:", cv_scores.std())


    # predict the test set
    y_pred = model.predict(X_test)

    print("\nTest Set Evaluation:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

   
    return end_time - start_time, accuracy_score(y_test, y_pred), cv_scores.mean()

# execute the function
training_time, test_accuracy, cv_accuracy = time_it()

print(f"\nSummary:")
print(f"Training Time: {training_time:.4f} seconds")
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"CV Mean Accuracy: {cv_accuracy:.4f}")
```
